# The correct version



## Program logic

#### The program runs inside a "while True"  loop
#### It takes a number from the user 
#### If the number is **Negative** the program stops
#### If the number is **Even** the loop continues
#### If the number is **odd** detailed information is printed

##### ______________________________________________________________________________


### features

##### for **odd numbers** , the program displays:

##### _ The number it self
##### _ whether the number is odd or even
##### _ The number of digits
##### _ whether the number is multiple of 5

##### _____________________________________________________________________________


#### Example output

##### .... Text
##### Enter a number : 15
##### Number : 15
##### Type : odd
##### Digits count : 2
##### Multiple of 5 : Yes

###### ___________________________________________________________________________________________________




```python
while True:
    number = int(input("Enter a number: "))

    #If the number is negative , exit the loop
    if number < 0:
        print("Number is negative , Program stoped!")
        break

    #If the number is even , the loop whould continue
    if number % 2 == 0:
        print("Number is even , continue...")
        continue

    #If the number is odd
    else:
        print("number:" , number)
        print("Type: odd")


    #Digits count
    digits_count = len(str(abs(number)))
    print("digits_count:" , digits_count)


    #Number multiple of five
    if number % 5 == 0:
        print("Multiple of 5: Yes")
    else:
        print("Multiple of 5: No")

    print("-" * 30)
    
```
